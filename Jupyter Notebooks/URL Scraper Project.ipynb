Creating a function to iterate through recipe type to generate the complete URL 

```python
def run(base_url, recipe_types):
    for recipe_type in recipe_types:
        counter = 1
        while True:
            endpoint = base_url+recipe_type
            if counter > 1:
                endpoint += f'/page/{counter}'
            #print(endpoint)
            result = requests.get(endpoint, headers = headers)
            if result.status_code != 200:
                break
            get_urls(result)
            #print(endpoint)
            #print(result.text)
            counter += 1
```
